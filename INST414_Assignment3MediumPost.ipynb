```python
import numpy as np
import pandas as pd
```

```python
fortnite_data = pd.read_csv('Fortnite_players_stats.csv')
fortnite_data.head()
```

```python
fortnite_data.info()
```

```python
fortnite_data = fortnite_data.drop(columns=['Duos score', 'Duos top1', 'Duos kd', 'Duos winRatio', 'Duos matches', 'Duos kills', 'Duos minutesPlayed','Trios score', 'Trios top1', 'Trios kd', 'Trios winRatio', 'Trios matches', 'Trios kills', 'Trios minutesPlayed',   'Squads score', 'Squads top1', 'Squads kd', 'Squads winRatio', 'Squads matches', 'Squads kills', 'Squads minutesPlayed',  'LTM score', 'LTM top1', 'LTM kd', 'LTM winRatio', 'LTM matches', 'LTM kills', 'LTM minutesPlayed', 'LTM top3'])
```

```python
fortnite_model = fortnite_data.set_index('Player')
fortnite_model.head(100)
```

```python
from scipy.sparse import lil_matrix
import scipy.spatial.distance
```

# PLAYER: Idk_Pi

```python
player_1 = 'Idk_Pi'
player_data = fortnite_model.loc[player_1]
distances = scipy.spatial.distance.cdist(fortnite_model, [player_data], metric = "euclidean")[:,0]
query_distances = list(zip(fortnite_data.index, distances))
```

```python
for sim_player, solo_stats in sorted(query_distances, key = lambda x: x[1], reverse = False)[:11]:
    print(fortnite_data['Player'].iloc[sim_player], solo_stats)
```
